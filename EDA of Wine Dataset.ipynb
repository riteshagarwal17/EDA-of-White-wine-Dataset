# EXPLORATORY DATA ANALYSIS ON WHITE WINE DATASET
#### INTRODUCTION OF DATA SET
White wine is a wine whose colour can be straw-yellow, yellow-green, or yellow-gold. It is produced by the alcoholic fermentation of the non-coloured pulp of grapes, which may have a skin of any colour.

##### OBJECTIVE:
The project is aimed at looking and finding deep insights of the white wine dataset.The purpose of this project is to determine what physical and chemical properties affect white wine quality through exploratory data analysis of a dataset containing 4898 white variants of the portuguese "vinho verde" wine.

Due to privacy and logistic issues only physical and chemical inputs are available.

###### DATA DESCRIPTION:
The white wine dataset have all physical and chemical input variables which effects quality of wine.The dataset have 12 variables,11 of are pyhsical and chemical properties and other input variable is of quality. All chemical properties data are continuous variable. Quality is an oridinal variable with possible rankings from 1(worst) to 10(best).

Dataset have 4898 rows and 12 columns respectively. The total number of values in dataset are 58776.

Different samples of white wine are tasted by differnt consumers and final rating is assigned to wine sample.

# BIVARIATE ANALYSIS
Bivariate analysis is one of the simplest forms of statistical analysis. it involves the analysis of two variables(often denoted as X,Y) for the purpose to determine the relationship b/w different quantities.

'Correlation' refers to a mutual relationship or association b/w differnt quantites.
Here we calculate correlation i.e statistical parameter of dataset which gives us relatioship b/w differnt quantities.

```python
df.corr()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(),annot=True)
sns.despine()
```

#### ALCOHOL Vs QUALITY PLOT
Here's jitter plot for alcohol Vs quality.


```python
sns.regplot(x='quality',y='alcohol',data=df,x_jitter=0.3,scatter_kws={'alpha':0.3},color="blue")
sns.despine()
plt.grid()
plt.title("ALCOHOL Vs QUALITY PLOT")
```

Following trends are observed from above alcohol Vs quality plot:

Alcohol and quality have correlation coefficient of 0.44
Above plot shows increasing trend b/w alcohol and quality i.e.better quality wines have more alcohol content
We can observe that mediocre quality wines i.e of quality (5 and 6) have alcohol content 9%-12% and 9%-14% by volume respectively

#### ALCOHOL Vs DENSITY PLOT

## MULTIVARIATE ANALYSIS
Multivariate data analysisn refers to any statistical technique used to analyze data that arises from more than one variable,and this techniques are very valuable when working with correlated variables.

FREE SULPHUR DIOXIDE Vs TOTAL SULPHUR DIOXIDE
Correlation coefficient of free suplhur dioxide and total sulphur dioxide is 0.62 which is relatively high.The content of both free sulphur dioxide and total sulphur dioxide effects the content of alcohol as corr coefficient of alcohol with two chemical propertie are -0.31 and -0.43 respectively. Therefore both properties efect alcohol content ,alcohol is directly corelated with quality. That's why this plot is impostant for analysing.


```python
sns.regplot(x='free.sulfur.dioxide' ,y='total.sulfur.dioxide',data=df,x_jitter=0.3,scatter_kws={'alpha':0.3})
sns.despine()
plt.grid()
plt.title("FREE SULPHUR DIOXIDE Vs TOTAL SULPHUR DIOXIDE PLOT")
```
